```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.set()
```

### Tips data

```
df = sns.load_dataset('tips')
df.head()
```

```
sns.set(style = 'ticks', palette = 'pastel')
sns.catplot(data = df, x = 'day', y = 'total_bill')
```

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'swarm')
```

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'box')
```

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'violin')
```

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'boxen')
```

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'point')
```

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'bar')
```

```
sns.catplot(data = df, x = 'day', kind = 'count')
```

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'swarm', hue = 'sex')
```

```
sns.catplot(data = df.query("day == ['Fri', 'Sat', 'Sun']"), x = 'day', y = 'total_bill', kind = 'swarm', hue = 'sex')
```

### Titanic data

```
df = sns.load_dataset('titanic')
df.head()
```

```
df.info()
```

```
df.describe().T
```

```
sns.catplot(data = df, x = 'sex', y = 'survived', kind = 'bar')
```

```
sns.catplot(data = df, x = 'sex', y = 'survived', kind = 'bar', hue = 'class')
```

```
sns.catplot(data = df, x = 'deck', kind = 'count', palette = 'Blues')
```

```
sns.catplot(data = df, y = 'deck', kind = 'count', palette = 'Blues')
```
