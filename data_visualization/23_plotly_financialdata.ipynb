```
def fetch_financial_data(company = 'AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name = company, data_source = 'stooq')

df = fetch_financial_data()
df = df.reset_index()
df.head()
```

```
df = df[df['Date'] > '2020-01-01']
df.describe().T
```

```
import plotly.graph_objects as go
```

### OHLC

```
fig = go.Figure(go.Ohlc(x = df['Date'],
                        open = df['Open'],
                        high = df['High'],
                        low = df['Low'],
                        close = df['Close']))
fig.show()
```

### Wykres świecowy

```
fig = go.Figure(go.Candlestick(x = df['Date'],
                               open = df['Open'],
                               high = df['High'],
                               low = df['Low'],
                               close = df['Close']))
fig.show()
```

### Waterfall

```
fig = go.Figure(go.Waterfall(x = ['Sprzedaż Stacjonarna', 'Sprzedaż online',
                                  'Przychody', 'Koszty', 'Podatek dochodowy',
                                  'Zysk netto'],
                             y = [60, 80, 0, -50, -30, 0],
                             measure = ['relative', 'relative', 'total',
                                        'relative', 'relative', 'total'],
                             text = ['+60', '+80', '', '-50', '-30', ''],
                             textposition = 'outside',
                             name = 'Rachunek zysków i strat 2019'))
fig.update_layout(title = 'Rachunek zysków i strat 2019',
                  showlegend = True)
fig.show()
```
